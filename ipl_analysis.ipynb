#### TOSS DECISION

```python
decision= ipl_df.groupby(ipl_df.toss_decision).toss_decision.count()
print(decision.to_string())
total=ipl_df.toss_decision.count()
print("total= ",total)
```

```python
matplotlib.rcParams['figure.figsize']=(9,9)
decision.plot.bar(color=['C8','C3']);
plt.title("Preferred choice after winning the toss: ")
plt.show()
```

The above bar plot gives us an idea that out of `756` matches that were played in `463` matches, the toss winning team chose fielding and in only `293` matches, the team decided to defend their score, i.e, chose to bat first. So, the team that has won the toss would most likely choose to chase the score.

### Q1. Does winning the toss means winning the match ? Find how many teams has chosen to chase and have won. Similarily, how many teams have won while defending the match.

```python
ipl_df[['team1','team2','toss_winner','toss_decision','winner']]
```

```python
df=ipl_df[ipl_df.toss_winner==ipl_df.winner]
df
```

```python
total=df.toss_decision.count()
field_df= df[df.toss_decision== 'field']
bat_df= df[df.toss_decision== 'bat']
```

```python
field_df[['team1','team2','toss_winner','toss_decision','winner']]
```

```python
bat_df[['team1','team2','toss_winner','toss_decision','winner']]
```

```python
win_pct= df.toss_decision.count() *100 / ipl_df.toss_decision.count()
field_pct= field_df.toss_decision.count() * 100 / total
bat_pct= bat_df.toss_decision.count() * 100 / total
print("Winning %age if you win the toss= {0:.2f}".format(win_pct))
print("Winning %age if you chose fielding= {0:.2f}".format(field_pct))
print("Winning %age if you chose batting= {0:.2f}".format(bat_pct))
```

There is a `47.35%` chance that if you win the toss you will win the match. Out of this, `65.64%` of the team won because they chose fielding and only `34.36%` won with the decision to bat first.                                                           So, it is safe to say that **winning the toss does not necessarily mean that you will win the match**.If you win the toss choose fielding to increase your winning chances.
